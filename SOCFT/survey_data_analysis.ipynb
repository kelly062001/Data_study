```python
import numpy as np
import pandas as pd
```

```python
survey_data = pd.read_csv('./survey_data_8.19.csv')
df=pd.DataFrame(survey_data)
df
```

```python
df['frequency_findinginfo'].unique()
```

```python
def frequency_findinginfo(x):
    if x == "사건에 대해 관심을 가지고 사건의 경위, 현재 상황 등 적극적으로 찾아보았다.":
        return 3
    elif x == "기사 제목만 읽어보는 등 대략적으로 찾아보았다.":
        return 2
    elif x=="사건에 대해 들었지만 찾아보지는 않았다.":
        return 1
    elif x=="칼부림 사건에 대해 들어본 적이 없다.":
        return 0
```

```python
df_new["frequency_findinginfo"] = df["frequency_findinginfo"].apply(frequency_findinginfo)
df_new
```

```python
def change_plan(x):
    if x=="일정을 변경한 적이 있다.(4-1로 이동)":
        return 1
    elif x=="예고 자체를 찾아보지 않는다.(다음 섹션으로 이동)":
        return 0
```

```python
df_new["change_plan"] = df["change_plan"].apply(change_plan)
df_new
```

```python
df['job'].unique()
```

```python
def job(x):
    if x=="초/중/고등학생":
        return 0
    elif x=="대학(원)생":
        return 1
    elif x=="직장인/자영업/프리랜서/전문직/공무원":
        return 2
    elif x=="전업주부":
        return 3
    elif x=="무직/취업준비생":
        return 4
    elif x=="기타":
        return 5
```

```python
import matplotlib

import seaborn as sns
sns.set(style='whitegrid')
import matplotlib.pyplot as plt # 맷플롯립 import 하기
plt.rcParams['font.family'] = 'NanumBarunGothic' # 나눔바른고딕 적용하기
matplotlib.rcParams['axes.unicode_minus'] = False
```
