```python
### 연습문제 1: 와인 데이터셋을 불러와 데이터프레임으로 만들어보세요.

### 연습문제 2: 기술통계 정보를 출력해보세요.
print(df.describe())

```

```python
### 연습문제 3: 와인의 종류(quality)별로 데이터가 몇 개씩 있는지 출력해보세요.
df['quality'].groupby(df['quality']).mean()
```

```python
### 연습문제 7: 각 와인의 타입(type)별로 평균 알코올 도수(alcohol)와 평균 당도(residual sugar)를 구해보세요.
df[df['type']=='red'][['alcohol','residual sugar']].mean()

```

```python
### 연습문제 8: 'quality' 열의 값이 5 이상인 데이터만 추출하시오.
df[df['quality']>=5]
```

```python
### 연습문제 9: 'fixed acidity' 열과 'pH' 열의 값이 각각 3 이상, 4이상 인 데이터만 추출하여 데이터 프레임을 생성하시오.

df[(df['fixed acidity'] >= 3) & (df['pH'] >= 4)]
```

```python
### 연습문제 10: 'volatile acidity' 열의 값이 0.5 이하인 데이터의 'quality' 열의 평균을 계산하세요.
filtered_df = df[df['volatile acidity'] <= 0.5]['quality'].mean()
filtered_df
```

```python
### 연습문제 11: 'residual sugar' 열의 값이 2 이하인 데이터의 'density' 열의 중앙값을 계산하세요.
df[df['residual sugar'] <= 2]['density'].mean()
```

```python
### 연습문제 12: 'alcohol' 열이 가장 높은 5개 데이터의 'quality' 열 값을 출력하세요.
df['alcohol'].sort_values().tail(5)
```

```python
### 연습문제 13: 'chlorides' 열의 값이 가장 높은 데이터 10개의 'pH' 열 값을 출력하세요.
### 연습문제 14: 'sulphates' 열의 값이 0.4 이하인 데이터의 'quality' 열 값을 1로 변경하세요.
### 연습문제 15: 'fixed acidity' 열의 값이 7 이하인 데이터의 'quality' 열 값을 0으로, 8 이상인 데이터의 'quality' 열 값을 2로 변경하세요. 나머지 데이터는 변경하지 않습니다.
### 연습문제 16: 'quality' 열의 값이 0인 데이터 중 'fixed acidity' 열의 값을 7로 변경하세요.
### 연습문제 17: 'quality' 열의 값이 6 이하인 데이터 중 'pH' 열의 값을 3으로 변경하세요.
```
