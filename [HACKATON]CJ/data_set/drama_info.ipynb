```python
genre_list=[]
for i in range(0, 753):
    k = korea_drama.loc[i, 'genre']
    k=k[1:-1]
    k=re.sub("[',]","", k)
    genre = list(map(str, k.split()))
    for i in genre:
        genre_list.append(i)
```

```python
for i in range(0, 753):
    try:
        end_date = datetime.strptime(korea_drama.loc[i, 'end_dt'],"%Y-%m-%d")
        start_date = datetime.strptime(korea_drama.loc[i, 'start_dt'],"%Y-%m-%d")
        korea_drama.loc[i, 'drama_period']= end_date-start_date
    except ValueError:
        korea_drama.loc[i, 'drama_period']= 'change'
```
