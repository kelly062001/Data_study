```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, ElementNotInteractableException
```

```python
actor_list = pd.read_csv('./actor_list_ver1_2.csv')
```

```python
actor_list
```

```python
driver = webdriver.Chrome()

actor_english_list = actor_list['actor_kor']

def actor_info(x):
    global actor_english_list
    name = x
    driver.get('http://naver.com')
    
    try:
        search = driver.find_element(By.ID, 'query')
        search_2 = search.send_keys(x)

        btn = driver.find_element(By.CLASS_NAME, 'btn_search').send_keys(Keys.ENTER)
        dramas = driver.find_element(By.XPATH, "//*[@id='main_pack']/section[1]/div[2]/div[2]/div/div[1]/a").send_keys(Keys.ENTER)
        infos = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='main_pack']/section[1]/div[2]/div/div/div[1]/div[3]/div/div/ul/li[1]/a/span/span")))
        info_text = infos.text
        
    except ElementNotInteractableException:
        dramas = driver.find_element(By.XPATH, "//*[@id='main_pack']/section[1]/div[2]/div[3]/div/div[1]/a[1]").send_keys(Keys.ENTER)
        try:
            infos = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='main_pack']/section[1]/div[2]/div/div/div[1]/div[3]/div/div/ul/li[1]/a/span/span")))
            info_text = infos.text
        except ElementNotInteractableException:
            infos = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='main_pack']/section[1]/div[2]/div/div/div[1]/h3/span")))
            info_text = infos.text
        except Exception as e:
            info_text = 'Notfound'
        
    except Exception as e:
        info_text = 'Notfound'
        
    return info_text


numbers_list =[]

for actor in actor_english_list:
    time.sleep(3)
    k = actor_info(actor)
    numbers_list.append(k)
    print(numbers_list)
print(numbers_list)

driver.quit()
```
