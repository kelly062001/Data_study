```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
```

```python
driver = webdriver.Chrome()

drama_eng_title = drama_list_2020['drama_name']

def english_to_korean(x):
    global drama_eng_title
    name = x
    url = f"https://mydramalist.com/search?q={name.replace(' ', '+')}&adv=titles&ty=68,83&so=relevance"
    driver.get(url)

    try: 
        dramas = driver.find_element(By.XPATH, "//*[@id]/div/div/div[2]/h6/a[1]").send_keys(Keys.ENTER)
        title = WebDriverWait(driver,5).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='show-detailsxx']/div[4]/ul[1]/li[1]/a")))
        title_text = title.text
    except TimeoutException as ex:
        title = WebDriverWait(driver,5).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='show-detailsxx']/div[4]/ul[1]/li[2]/a")))
        title_text = title.text
        
    except Exception as e:
        title_text = "Notfound"
        
    return title_text
# 결과를 저장할 빈 리스트 생성
drama_title_2020 = []

i=0
# 영문 제목 번역 
for title in drama_eng_title:
    time.sleep(3)
    korean_title = english_to_korean(title)
    drama_title_2020.append(korean_title)
    print(drama_title_2020)
    print(i)
    i+=1

# 번역 결과
drama_list_2020['kor_title'] = drama_title_2020

driver.quit()
```
