```python
import pandas as pd

reviews = pd.read_csv("./reviews.csv")
korean_drama = pd.read_csv("./korean_drama.csv")
recommendations = pd.read_csv("./recommendations.csv")
wiki_actors = pd.read_csv("./wiki_actors.csv")
korean_drama = korean_drama.rename(columns={'drama_name': 'title'})
recommendations = recommendations.rename(columns={'kdrama_name': 'title'})
wiki_actors = wiki_actors.rename(columns={'drama_name': 'title'})

merged_data = reviews.merge(korean_drama, on='title', how='inner')
merged_data = merged_data.merge(recommendations, on='title', how='inner')
merged_data = merged_data.merge(wiki_actors, on='title', how='inner')
merged_data.to_csv('./merged_file.csv', index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException




```

```python
# 웹 드라이버 초기화
driver = webdriver.Chrome()

drama_eng_title = korean_drama_2020['title']

def english_to_korean(x):
    global drama_eng_title
    name = x
    url = f"https://mydramalist.com/search?q={name.replace(' ', '+')}"
    driver.get(url)

    try: 
        dramas = driver.find_element(By.XPATH, "//*[@id]/div/div/div[2]/h6/a[1]").send_keys(Keys.ENTER)

        title = WebDriverWait(driver,20).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='show-detailsxx']/div[4]/ul[1]/li[1]/a")))
        
        title_text = title.text
    except TimeoutException as ex:
        title_text = "Notfound"
        
    return title_text
# 결과를 저장할 빈 리스트 생성
drama_kor_title = []

i=0
# 영문 제목 번역 
for title in drama_eng_title:
    time.sleep(3)
    korean_title = english_to_korean(title)
    drama_kor_title.append(korean_title)
    print(i, title, korean_title)
    i+=1

# 번역 결과
korean_drama_2020['korean_title'] = drama_kor_title

driver.quit()
```
