```python
import numpy as np
import matplotlib.pyplot as plt
X = np.linspace(0.5, 50,100)
from scipy.stats import chi2
plt.figure(figsize=(10, 6))
for df in [1,2,3,4,5]:
    plt.plot(X, chi2(df).pdf(X), label=df)
    plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(X, chi2(1).pdf(X), label=1)
plt.legend()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(X, chi2(3).pdf(X), label=3)
plt.legend()
```
