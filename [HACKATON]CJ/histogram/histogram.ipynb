```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
drama_list = pd.read_csv('./new_drama_info_ver3.csv')
```

```python
pd.set_option('display.max_columns', None)
drama_list
```

```python
conditions = [ 
    (drama_list['month'] == 12) | (drama_list['month'] <= 2),
    (drama_list['month'] >= 3) & (drama_list['month'] <= 5),
    (drama_list['month'] >= 6) & (drama_list['month'] <= 8),
    (drama_list['month'] >= 9) & (drama_list['month'] <= 11)
    ]
values = ['Winter', 'Spring', 'Summer', 'Autumn']
drama_list['season'] = np.select(conditions, values)
drama_list.head()
```

```python

#season
plt.hist(drama_list['season'], bins=20, color='skyblue', edgecolor='k')
plt.xlabel('Season')
plt.ylabel('Frequency')
plt.title('Histogram of Season')
```
