```python
drama_original_list = pd.read_csv('./drama_info_ver6.csv')
```

```python
drama_original_list
drama_original = drama_original_list.copy()
drama_original
```

```python
drama_original['director']=drama_original_list['director'].fillna(0)
```

```python


for i in range(0,1752):
    if drama_original.loc[i,'director']==0:
        drama_title = drama_original.loc[i,'drama_name']
        print(drama_title)
        try:
            drama_director = drama_list.loc[drama_list['drama_name']==drama_title, 'director'][0]
            print(drama_director)
            drama_original.loc[i,'director']=drama_director
            print(drama_director)
        except Exception:
            pass
```

```python
drama_original['director'].value_counts()
```

```python
drama_original['screenwriter']=drama_original_list['screenwriter'].fillna(0)
```

```python
drama_original
```

```python
for i in range(0,1752):
    if drama_original.loc[i,'screenwriter']==0:
        drama_title = drama_original.loc[i,'drama_name']
        print(drama_title)
        try:
            drama_screenwriter = drama_list.loc[drama_list['drama_name']==drama_title, 'screenwriter'][0]
            print(drama_screenwriter)
            drama_original.loc[i,'screenwriter']=drama_screenwriter
            print(drama_screenwriter)
        except Exception:
            pass
```
