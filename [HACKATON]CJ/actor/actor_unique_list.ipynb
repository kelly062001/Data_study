```python
actor_list_blank=[]

for actor in actor_name:
    try: 
        k=actor[1:-1]
        new_name = re.sub("'","",k)
        new_name_list =list(map(str, new_name.split(', ')))
        for name in new_name_list:
            actor_list_blank.append(name)
    except Exception:
        pass

print(actor_list_blank)
```

```python
from collections import Counter
```

```python
counter = Counter(actor_list_blank)
```

```python
counter
```

```python
df = pd.DataFrame()

name_list=[]
drama_num_list=[]
for key, value in counter.items():
    name_list.append(key)
    drama_num_list.append(value)
```

```python
len(name_list)
```

```python
len(drama_num_list)
```

```python
df['actor_name']=name_list
df['drama_num']=drama_num_list
```
