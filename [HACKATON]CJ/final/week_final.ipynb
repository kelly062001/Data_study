```python
import pandas as pd
from collections import Counter
```

```python

drama_list = pd.read_csv('./ver2_org_net_preprocessing.csv')
```

```python
#1700개 짜리 처음 데이터
director = pd.read_csv('./drama_info_ver6.csv')
```

```python
director
```

```python
import re
director_list_blank=[]
director_list = director['director']

for director_indivisual in director_list:
    try: 
        k=director_indivisual[1:-1]
        new_name = re.sub("'","",k)
        new_name_list =list(map(str, new_name.split(', ')))
        for name in new_name_list:
            director_list_blank.append(name)
    except Exception:
        pass

print(director_list_blank)
```

```python
director_counter = Counter(director_list_blank)
```

```python
director_counter
```

```python
director_drama_num = 0 
director_list = drama_list['director']
director_drama_max = []
i=0


for director_indivisual in director_list:
    finding_max=[]
    try:
        k=director_indivisual[1:-1]
        new_name = re.sub("'","",k)
        new_name_list =list(map(str, new_name.split(', ')))
        i+=1
        for name in new_name_list:
            print(name)
            finding_max.append(director_counter[name])
        director_drama_max.append(max(finding_max))
    except Exception :
        director_drama_max.append('NaN')
    
        
```

```python
len(director_drama_max)
```

```python
director_drama_max
```

```python
drama_list
```

```python
drama_list['director_drama_num']=director_drama_max
```

```python
drama_list
```

```python
writer_list_blank=[]
writer_list = drama_list['screenwriter']

for writer_indivisual in writer_list:
    try: 
        k=writer_indivisual[1:-1]
        new_name = re.sub("'","",k)
        new_name_list =list(map(str, new_name.split(', ')))
        for name in new_name_list:
            writer_list_blank.append(name)
    except Exception:
        pass

print(writer_list_blank)
```

```python
screenwriter_counter=Counter(writer_list_blank)
```

```python
screenwriter_counter
```

```python
director_drama_num = 0 
screenwriter_list = drama_list['screenwriter']
screenwriter_drama_max = []
i=0

for screenwriter_indivisual in screenwriter_list:
    finding_max=[]
    try:
        k=screenwriter_indivisual[1:-1]
        new_name = re.sub("'","",k)
        new_name_list =list(map(str, new_name.split(', ')))
        i+=1
        for name in new_name_list:
            print(name)
            finding_max.append(screenwriter_counter[name])
        screenwriter_drama_max.append(max(finding_max))
    except Exception :
        screenwriter_drama_max.append('NaN')
```

```python
screenwriter_drama_max
```

```python
drama_list['screenwriter_drama_num']=screenwriter_drama_max
```

```python
drama_list.loc[drama_list['director_drama_num'] != drama_list['director_drama_num'], 'director_drama_num'] = 1
drama_list.loc[drama_list['screenwriter_drama_num'] != drama_list['screenwriter_drama_num'], 'screenwriter_drama_num'] = 1
```

```python
drama_list.to_csv('./drama_info_final_ver_1.csv', index=False,encoding="utf-8-sig" )
```
